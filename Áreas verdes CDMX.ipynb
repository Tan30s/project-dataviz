#### De la siguiente pagina de [inventario de areas verdes de la CDMX]( https://sedema.cdmx.gob.mx/programas/programa/inventario-de-areas-verdes) obtuvimos una tabla que nos indica el número de superficie por habitante 

```python
url="https://sedema.cdmx.gob.mx/programas/programa/inventario-de-areas-verdes"
html=requests.get(url).content
soup=BeautifulSoup(html,"lxml")
tablas=pd.read_html(html,encoding = 'utf8')
tabla=tablas[1]
tabla
```

```python
tabla.columns
```

```python
tabla.columns=["Clave","Alcaldía", "Habitantes (Año 2015)", "Superficie (m²)", "Superficie por habitante (m²)"]
tabla
```

 De acuerdo a la Organización Mundial de la Salud (OMS), se requieren de mínimo 9m2 de áreas verdes por habitante para garantizar su bienestar este dato se obtuvo del siguiente [link](https://www.forbes.com.mx/cdmx-carece-de-espacios-verdes-suficientes-greenpeace/)

```python
tabla = tabla.iloc[1:]
```

```python
tabla.to_csv("Data sets/sup_x_hab.csv")
```

Según la siguiente [pagina](https://ibero.mx/prensa/estudio-en-ocho-anos-cdmx-perdio-187-km2-de-areas-verdes-centrus?fbclid=IwAR2L_z-nb4Ns99amM2wK9g_zdnPlq6PA7dLuZMFyLRb0hLeLiGCSzFyXhLk) la superficie urbana de la CDMX en el año 2017 fue de **610.5 km²**. Por otro lado la pagina del [inegi](http://cuentame.inegi.org.mx/monografias/informacion/df/territorio/default.aspx) indica que la superficie total de la CDMX es de **1494.3 km²** calcularemos que porcentaje de superficie representa el área urbana en la CDMX y de ella que porcentaje abarcan las áreas verdes urbanas de la CDMX

```python
# En metros cuadrados la superficie de superficie urbana es de:
sup_urb=610.5*1000000
sup_urb
```

```python
#Superficie total den metros cuadrados de la CDMX 
sup_tot= 1494.3*1000000
sup_tot
```

```python
#Este es el porcentaje de superficie que abarcan las áreas urbanas 
sup_urb*100/sup_tot
```

```python
#Superficie de areas verdes urbanas en metros cuadrados en CDMX
sup_avu=data.superficie.sum()
sup_avu
```

```python
#Porcentaje de superficie de areas verdes en zonas urbanas 
sup_avu*100/sup_urb
```
